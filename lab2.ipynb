```python
def q3():
    n1 = int(input('Enter the first number: '))
    n2 = int(input('Enter the second number: '))

    if n1 == n2:
        # checking for string equality
        print('The two numbers are the same')
    else: # nested conditional statement
        # only if not equal, then fo to the else part, and make another decision
        if n1 > n2:
            print('First number is bigger')
        else:
            print('Second number is bigger')

q3()
```

```python
def q3(): # 3-way selection statement
    n1 = int(input('Enter the first number: '))
    n2 = int(input('Enter the second number: '))

    if n1 == n2:
        # checking for string equality
        print('The two numbers are the same')
    elif n1 > n2:
        print('First number is bigger')
    else:
        print('Second number is bigger')

q3()
```

```python
def lab1q4():
    # read the number of seconds
    n = int(input("Enter time duration in seconds: "))

    # extract h, m, s
    h = n // 3600 # 3600 seconds is 1 hour
    m = n % 3600 // 60  # n // 60 % 60
    s = n % 60

    # print sum and product
    print(f'{n} seconds is {h} hour{"s" if h > 1 else ""}, '\
     +f'{m} minute{"s" if m > 1 else ""} and {s} second{"s" if s > 1 else ""}')

lab1q4()
```

```python
def q7():
    nric = input("Enter NRIC: ") #S1234567A

    if len(nric) != 9:
        print('Length must be exactly 9')
    elif nric[0].upper() not in "STFG":
        print('The first letter must be S, T, F or G')
    elif not nric[1:-1].isdigit():
        print('Must consist of 7 digits')
    elif not nric[-1].isalpha():
        print('Reference letter must be A to Z or a to z')
    else:
        print('Valid NRIC')
```

```python
def q9():
    op1, op, op2 = input('Enter arithmetic expression: ').split() # 23.6 + 33.2
    op1, op2 = float(op1), float(op2)
    # list unpacking ['23.6', '+', '33.2']

    # aList = input('Enter arithmetic expression: ').split() # 23.6 + 33.2
    # # list  ['23.6', '+', '33.2']
    # # op1, op, op1 = float(aList[0]), aList[1], float(aList[2])
    # aList[0], aList[1], aList[2] = float(aList[0]), aList[1], float(aList[2])
    # # list  [23.6, '+', 33.2]

    if op not in '+-/*': # in checks whether op is a member of a str collection '+-/*'
        # in checks whether op is a member of a tuple collection ('+', '-', '/', '*', '//')
        # in checks whether op is a member of a list collection ['+', '-', '/', '*']
        print('Invalid arithmetic operator')

    else: # op in '+-/*
        if op == '+':
            result = op1 + op2
        elif op == '-':
            result = op1 - op2
        elif op == '*':
            result = op1 * op2
        else:
            result = op1 / op2

        print(f'Result: {result}')

q9()
```
