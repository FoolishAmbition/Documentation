```python
def qn4():
  # reads a string and checks whether it is a palindrome
  user_input = input("Enter a word: ").lower()

  # flip the word to check if it is a palindrome
  flipped_word = user_input[::-1]

  if user_input == flipped_word:
    print(f"{user_input} is a palindrome")
  else:
    print(f"{user_input} is not a palindrome")

  return user_input, flipped_word

qn4()
```

```python
def qn9():
  '''
  reads 2 numbers in order to perform a simple arithmetic operation
  there is a total of 3 inputs, separated by an arithmetic operation
  in the following format float operator float
  where float is any decimal number and operator is either +,-,*,/
  between float, operation and float, there is a space between
  assume numeric values are valid, but operator may be invalid
  '''

  # reads float operator float and verify correct operator
  a = input("Enter arithmetic expression: ")
  n1, op, n2 = a.split()
  n1,n2 = map(float, (n1,n2))

  if op not in "+-*/": # checks whether op is a member of a collection
    # display "Invalid arithmetic operator" if operator is not within range
    print("Invalid arithmetic operator")
  else: # performs calculation of float operator float
    if op == "+":
      print(f"Result: {n1+n2:.2f}")
    elif op == "-":
      print(f"Result: {n1-n2:.2f}")
    elif op == "*":
      print(f"Result: {n1*n2:.2f}")
    else:
      print(f"Result: {n1/n2:.2f}")

qn9()
```

10. Write a triangle checker function, `question10()`, that reads in the 3 sides of a triangle. The function checks whether the 3 values can form a triangle. A triangle will not be possible if the sum of any two is less than or equal to the third. If the values can form a triangle, the function will display the type of triangle and property of the sides according to the table below.


    | Property of the sides    | Type of Triangle |
    | -------- | -------- |
    | All 3 sides are equal.   | Equilateral     |
    | Any 2 sides are equal. | Isosceles    |
    | All 3 sides are unequal. | Scalene    |


    Example runs:

    <u>Run 1</u>

    Side 1: **1**
    <br>Side 2: **1**
    <br>Side 3: **3**

            Not possible to form a triangle
        


    <u>Run 2</u>

    Side 1: **1**
    <br>Side 2: **1**
    <br>Side 3: **1**

            Triangle type: Equilateral (All 3 sides are equal)


```python
def qn10():
  # reads user input to form the sides of a triangle
  s1,s2,s3 = map(float,input("Three side of the triangle, "\
  "separated by a space: ").split())

  # checks whether the 3 values can form a triangle
  '''
  A triangle will not be possible if the sum of any two is less than
  or equal to the third
  '''

  if s1 + s2 <= s3 or s1 + s3 <= s2 or s2 + s3 <= s1:
    print("Not possible to form a triangle")
  else: # if values can form a triangle, print the type of triangle
    if s1 == s2 == s3:
      print("Triangle type: Equilateral (All 3 sides are equal)")
    elif s1 == s2 or s1 == s3 or s2 == s3:
      print("Triangle type: Isosceles (Any 2 sides are equal)")
    elif s1 != s2 != s3:
      print("Triangle type: Scalene (All 3 sides are unequal)")

  '''
  if all 3 sides are equal display "equilateral"
  if any 2 sides are equal display "isosceles"
  if all 3 sides are unequal display "scalene"
  '''


qn10()
```
