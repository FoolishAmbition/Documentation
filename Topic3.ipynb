# Indefinite loop - Structure 1

Don't know how many times we need to loop, no use to generate bean bags

while keyword

syntax:

```
initialise the var - initialisation statement
while continuing condition involves the var applies:
    loop body statements
    update the var in the loop condition to, at some point, make condition false - the update statement


the other statements not in the loop

```

````python
from random import choice
def q6b():
    rounds = int(input('How many rounds to play? : '))

    # accoumulator pattern
    correct = 0 # initialise the accumulator outside the loop

    # BEAN BAG PATTERN
    #for i in range(rounds): # start is 0
    for i in range(1, rounds+1): # start is 1
        toss = choice('HT')
        print(toss) # for checking our logic
        #guess = input(f'Round {i+1}: Head or Tail (H or T): ')
        guess = input(f'Round {i}: Head or Tail (H or T): ')

        if guess.upper() == toss:
            print('Correct')
            correct += 1
        else:
            print('Wrong')

    print(f'You guess {correct} correct out {rounds} rounds')

q6b()
````

````python
from random import choice
def q6c():
    # accumulator pattern
    countGuesses = 0

    toss = choice('HT') # initialisation statement
    print(toss) # for checking logic
    guess = input('Head or Tail (H or T): ') # initialisation statement
    countGuesses += 1

    while guess.upper() != toss: # continuing condition
        print('Wrong')

        toss = choice('HT') # update statement
        print(toss) # for checking logic
        guess = input('Head or Tail (H or T): ') # update statement
        countGuesses += 1

    print('Correct!')
    print(f'You got it in {countGuesses} tosses!')


q6c()

````

````python
from random import choice
def playOneGame():
    # accumulator pattern
    countGuesses = 0

    toss = choice('HT') # initialisation statement
    print(toss) # for checking logic
    guess = input('Head or Tail (H or T): ') # initialisation statement
    countGuesses += 1

    while guess.upper() != toss: # continuing condition
        print('Wrong')

        toss = choice('HT') # update statement
        print(toss) # for checking logic
        guess = input('Head or Tail (H or T): ') # update statement
        countGuesses += 1

    print('Correct!')
    print(f'You got it in {countGuesses} tosses!')

def q6d():
    play = 'Y'
    while play.upper() == 'Y':
        playOneGame()
        play = input('Play again? (y/n)')

    print('Program end')

q6d()
````

```
while True:

    loop statements
    var must be defined/given some values

        break # to break out from the loop the break statement is in

    loop statements
```



````python
def q3b():
    n = int(input("Enter number: ")) # initialisation step
    while n != -1: # continuing condition
        # generate bean bags from 1 to n, including n
        for i in range(1, n+1):
            print(f'{i:2} x {n} = {i*n:3}')

        n = int(input("Enter number: ")) # update step in the while loop

q3b()
````

````python
def q3b():

    while True:
        n = int(input("Enter number: ")) # first iteration is initialsation
         # subsequent iteration is update statement

        # check n before deciding whether to print time table
        if n == -1: # terminating condition
            break

        # generate bean bags from 1 to n, including n
        for i in range(1, n+1):
            print(f'{i:2} x {n} = {i*n:3}')
q3b()
````

````python
def q4():
    # accumulator pattern
    total = 0

    quantity = int(input("Enter quantity: ")) # init statement
    while quantity != -1: # conitnuing condition
        unitPrice = float(input('Enter unit price: '))
        subtotal = quantity * unitPrice

        # accumulator pattern
        total += subtotal

        print(f'Subtotal is ${subtotal:.2f}')

        quantity = int(input("Enter quantity: ")) # update statement

    print(f'Total price is ${total:.2f}')
    gst = 0.07 * total
    print(f'GST is ${gst:.2f}')
    print(f'Please pay ${total + gst:.2f}')

q4()
````

````python
def getValidValue(msg, x, y):
    value = float(input(f'{msg}: '))

    while value > y or value < x:
        print('Invalid input')
        value = float(input(f'{msg}: '))

    return value

mark = getValidValue('Enter mark', 1, 100)
print(f'The valid mark is {mark}')
````

````python
def divide(x, y):
    # accumulator pattern, count how many groups of y in x
    quotient = 0
    while x >= y: # continuing condition
        x -= y
        quotient += 1

    return quotient, x

top, bottom = 11, 4  # global variables are ok in text code but not in your running application
grp, rem = divide(top, bottom)
print(f'There are {grp} of {bottom} in {top} and the remainder is {rem}')
````
