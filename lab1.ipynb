Write a program that takes in a 3 digit integer and displays the sum and
product of the digits. E.g. if the number is 123, the sum displayed is 6 and the
product is also 6.

```python
def qn3():
  # read one 3-digit number

  digit = int(input("Please enter a 3 digit integer: "))

  # extract individual digits

  ones = digit % 10
  tens = digit % 100 // 10
  hundreds = digit // 100

  # compute the sum and product

  sum_ = hundreds + tens + ones
  product = hundreds * tens * ones

  print(f'The sum of the digits {hundreds}, {tens}, and {ones}, is {sum_} and ' +\
        f'the product is {product}')

qn3()
```

Write a program that reads in a positive integer representing time in seconds
and converts it to hour, minute and seconds. For example, if the input is 3670
seconds, output 1 hour, 1 minute and 10 seconds.

```python
def qn4():
  # read the seconds for input
  seconds = int(input("Please input the seconds to convert: "))

  # converting seconds into hour, minute, and seconds format
  hr = seconds // 3600 # floor division to find our the hours
  # find the remainder after getting hours, and find the exact minutes
  min = (seconds % 3600) // 60
  # find the remainders after minutes
  sec = (seconds % 3600) % 60

  # display the output
  return hr, min, sec

qn4()
```

5. Write a function, `question5()`, that reads an input representing a change which is an amount less than 1 dollar. The funcion calculates the change into 50, 10, 5, and 1 cent coins. The function then displays the number of each coin required for that change.

    For example:
    
    Enter change: **47**

        50 cent: 0
        10 cent: 4
        5 cent: 1
        1 cent: 2
    
    Print the above display output and return the values of the number of each coin for 50, 10, 5 and 1 cent coins."

```python
def question5():
  # reads input representing a change (amount less than 1 dollar)
  change = int(input("Please input the change: "))

  # calculates the change into 50, 10, 5 and 1 cent coins

  c50 = change // 50 # floor division to find out how many 50c we need
  c10 = (change % 50) // 10 # find the remainder after 50c, and then find 10c
  c5 = ((change % 50) % 10) // 5 # find remainder after 50c, and 10c and find 10c
  c1 = ((change % 50) % 10) % 5 # find the remainder final

  # display number of each coins required for change
  print("50 cent: ", c50)
  print("10 cent: ", c10)
  print("5 cent: ", c5)
  print("1 cent: ", c1)

  return c50, c10, c5, c1


question5()
```

6. A restaurant is offering meals at 50% discount.

    A service charge (10%) and GST (7%) apply to the discounted cost. While the service charge applies to the discounted price, note that the GST calculation is based on the total of the discounted amount and the service charge.

    Write a function, `question6()`, that reads in the cost of the meal and displays a detailed receipt. An example is as follows:

    Enter meal amount ($): **120**

       Receipt
       
       Cost of meal:   $120.00
       
       50% discount:   $ 60.00
       
       Service charge: $  6.00
       
       GST:            $  4.62
       
       Total amount:   $ 70.62
        
  > Output should be formatted.

  Print the above display output and return the exact values rounded to two decimal places of the cost of the meal, the discount, the service charge, the gst, and the total amount."

```python
def question6():
  # reads the cost of the meal
  cost = float(input("Please enter the cost of the meal: "))

  # applying 50% discount onto costs
  discount = cost * 0.5

  # applying service charge (10%) onto the 50% discount
  service_charge = discount * 0.1

  # applying GST (7%) on total discounted amount AND service charge
  gst = round((discount + service_charge) * 0.07,2)

  # Total amount need to pay
  total = round(discount + service_charge + gst,2)

  '''
  displaying of the everything

  in this format

         Receipt

       Cost of meal:   $120.00

       50% discount:   $ 60.00

       Service charge: $  6.00

       GST:            $  4.62

       Total amount:   $ 70.62


  '''
  print(f"""
  Receipt
  Cost of meal:   ${cost:6.2f}
  50% discount:   ${discount:6.2f}
  Service charge: ${service_charge:6.2f}
  GST:            ${gst:6.2f}
  Total amount:   ${total:6.2f}
  """)
  return cost, discount, service_charge, gst, total

question6()
```
